```python
import subprocess
import sys
import os

# Tell pytest NOT to create cache folders — fixes the Databricks error
os.environ["PYTHONDONTWRITEBYTECODE"] = "1"

TEST_DIR = "/Workspace/Users/dvmb1412@gmail.com/Traffic-Data-Pipeline/tests"

result = subprocess.run(
    [
        sys.executable, "-m", "pytest",
        TEST_DIR,
        "-v",
        "--tb=short",
        "--no-header",
        "-p", "no:cacheprovider",    # disables cache
        "-p", "no:faulthandler",     # disables fault handler
        "--import-mode=importlib",   # fixes workspace import issues
    ],
    capture_output=True,
    text=True,
    env={**os.environ, "PYTHONDONTWRITEBYTECODE": "1"}
)

print(result.stdout)
if result.stderr:
    print("ERRORS:", result.stderr[:3000])

print(f"\nExit code: {result.returncode}")
print("ALL PASSED ✓" if result.returncode == 0 else "SOME TESTS FAILED ✗")
```
